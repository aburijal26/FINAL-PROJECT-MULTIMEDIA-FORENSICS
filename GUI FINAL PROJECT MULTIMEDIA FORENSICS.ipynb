```python
import cv2
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics.pairwise import cosine_similarity
import tkinter as tk
from tkinter import filedialog, Label, Button

def compare_images():
    original_img = cv2.imread(original_path.get())
    forgery_img = cv2.imread(forgery_path.get())
    sift = cv2.xfeatures2d.SIFT_create()
    original_kp, original_desc = sift.detectAndCompute(original_img, None)
    forgery_kp, forgery_desc = sift.detectAndCompute(forgery_img, None)
    original_clustering = AgglomerativeClustering().fit(original_desc)
    original_clusters = original_clustering.labels_
    forgery_clustering = AgglomerativeClustering().fit(forgery_desc)
    forgery_clusters = forgery_clustering.labels_
    similarity = cosine_similarity(original_desc, forgery_desc)
    similarity = similarity[0][0]
    threshold = 0.8
    if similarity > threshold:
        result_label.config(text=('Kedua Citra sama, dengan Similarity', similarity))
    else:
        result_label.config(text=('Kedua Citra berbeda, dengan Similarity', similarity))
    original_label = Label(root, image=original_img)
    original_label.grid(row=3, column=0, columnspan=2)
    forgery_label = Label(root, image=forgery_img)
    forgery_label.grid(row=3, column=2, columnspan=2)

root = tk.Tk()
root.title("Region duplication detection in digital images based on Centroid Linkage Clustering of Key-points and graph similarity")

original_path = tk.StringVar()
forgery_path = tk.StringVar()

original_label = tk.Label(root, text="Original Image:")
original_label.grid(row=0, column=0, padx=10, pady=10)

original_entry = tk.Entry(root, textvariable=original_path)
original_entry.grid(row=0, column=1, padx=10, pady=10)

original_button = tk.Button(root, text="Browse", command=lambda: original_path.set(filedialog.askopenfilename()))
original_button.grid(row=0, column=2, padx=10, pady=10)

forgery_label = tk.Label(root, text="Forgery Image:")
forgery_label.grid(row=1, column=0, padx=10, pady=10)

forgery_entry = tk.Entry(root, textvariable=forgery_path)
forgery_entry.grid(row=1, column=1, padx=10, pady=10)

forgery_button = tk.Button(root, text="Browse", command=lambda: forgery_path.set(filedialog.askopenfilename()))
forgery_button.grid(row=1, column=2, padx=10, pady=10)

compare_button = tk.Button(root, text="Compare", command=compare_images)
compare_button.grid(row=2, column=0, padx=10, pady=10)

result_label = tk.Label(root, text="")
result_label.grid(row=2, column=1, columnspan=2, padx=10, pady=10)

root.mainloop()
```
